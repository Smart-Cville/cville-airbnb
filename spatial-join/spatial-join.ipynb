```python
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

import tqdm
from tqdm._tqdm_notebook import tqdm_notebook
```

```python
df = pd.read_csv("airbnb/airbnb_cville.csv")
df["geom"] = df[['latitude','longitude']].apply(lambda x : '{}, {}'.format(x[0],x[1]), axis=1)

with pd.option_context("display.max_columns", 50):
    display(df)
```

```python
locator = Nominatim(user_agent="myGeocoder", timeout=10)
rgeocode = RateLimiter(locator.reverse, min_delay_seconds=0.001)
```

```python
tqdm_notebook().pandas()

# rgeocode(df['geom'][0]).raw["address"]
df['address'] = df['geom'].progress_apply(rgeocode)
df.head()
```

It's likely that city data is in address format (not geographic coordinates).

A couple options to join AirBnB data to City data:
1. Reverse-geocoding - Convert AirBnB coordinates to address & join on address
    - OpenStreetMap geocoder is free, but for many rows, it doesn't get more detailed than street name
    - Other geocoders (i.e. GoogleV3) may be better, but I think you have to pay for these?
2. Geocoding - Convert City address to coordinates & join on coordinates 
    - Will need to add a buffer & join city data to AirBnB data if points are within a small distance (25 meters?)
    - Will be an issue with apartments or properties really close together
